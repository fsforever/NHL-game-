```python
import requests
import zipfile
import os
```

```python
# Retrieve NHL dataset
url = 'https://www.kaggle.com/api/v1/datasets/download/martinellis/nhl-game-data'

response = requests.get(url)

if response.status_code == 200:
    with open('/lakehouse/default/Files/nhl-dataset.zip', 'wb') as file:
        file.write(response.content)
else:
    print(response.status_code)
```

```python
# Extract zip file
zip_file_path = '/lakehouse/default/Files/nhl-dataset.zip'
to_dir = '/lakehouse/default/Files/dataset'

with zipfile.ZipFile(zip_file_path, 'r') as zip:
    zip.extractall(to_dir)
```
