```python
# Portfolio-ready script: Download and extract NHL dataset

import requests
import zipfile
import os

# ----------------------------
# Step 1: Define dataset URL and paths
# ----------------------------
dataset_url = 'https://www.kaggle.com/api/v1/datasets/download/martinellis/nhl-game-data'
zip_file_path = '/lakehouse/default/Files/nhl-dataset.zip'
extracted_dir = '/lakehouse/default/Files/dataset'

# Ensure extraction directory exists
os.makedirs(extracted_dir, exist_ok=True)

# ----------------------------
# Step 2: Download the dataset
# ----------------------------
try:
    print("Downloading NHL dataset...")
    response = requests.get(dataset_url, stream=True)
    
    if response.status_code == 200:
        with open(zip_file_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print(f"Dataset successfully downloaded to {zip_file_path}")
    else:
        print(f"Failed to download dataset. Status code: {response.status_code}")

except Exception as e:
    print(f"An error occurred during download: {e}")

# ----------------------------
# Step 3: Extract the dataset
# ----------------------------
try:
    print("Extracting dataset...")
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extracted_dir)
    print(f"Dataset successfully extracted to {extracted_dir}")

except FileNotFoundError:
    print("Zip file not found. Please check the download step.")
except zipfile.BadZipFile:
    print("Error: Bad zip file. Download may have failed.")
except Exception as e:
    print(f"An error occurred during extraction: {e}")
```
